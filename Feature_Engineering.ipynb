```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
df = pd.read_csv("Train.csv")
```

```python
# Copy the dataset to avoid changing original
data = df.copy()
```

```python
# Create new feature: Item Category
data['Item_Category'] = data['Item_Identifier'].apply(lambda x: x[:2])
data['Item_Category'] = data['Item_Category'].map({'FD': 'Food','NC': 'Non-Consumable','DR': 'Drinks'})
```

```python
# Mark Non-Consumables as 'None' in Fat Content
data.loc[data['Item_Category'] == 'Non-Consumable', 'Item_Fat_Content'] = 'None'
```

```python
# Label Encoding for ordinal variables
le = LabelEncoder()
data['Outlet'] = le.fit_transform(data['Outlet_Identifier'])
```

```python
# One-hot encoding for nominal categorical columns
cat_cols = ['Item_Fat_Content', 'Item_Type', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'Item_Category']
data = pd.get_dummies(data, columns=cat_cols)
```

```python
# Drop unnecessary columns
data.drop(['Item_Identifier', 'Outlet_Identifier', 'Outlet_Establishment_Year'], axis=1, inplace=True)
```
