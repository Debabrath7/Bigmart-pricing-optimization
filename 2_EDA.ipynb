```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("Train.csv")
```

```python
print("Train Columns:\n", df.columns.tolist())
```

```python
# Set visual style
sns.set(style="whitegrid")
```

```python
# Sample subset to speed up heavy plots
eda_df = df.copy()
```

```python
# Plot 1: Distribution of Item Outlet Sales
plt.figure(figsize=(9, 5))
sns.histplot(df['Item_Outlet_Sales'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Item Outlet Sales')
plt.xlabel('Sales')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Distribution of Item MRP
plt.figure(figsize=(9, 5))
sns.histplot(eda_df['Item_MRP'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Item MRP (Price)')
plt.xlabel('Item MRP')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Item MRP vs Item Outlet Sales (scatter plot)
plt.figure(figsize=(9, 5))
sns.scatterplot(data=eda_df, x='Item_MRP', y='Item_Outlet_Sales', alpha=0.5)
plt.title('Item MRP vs Item Outlet Sales')
plt.xlabel('Item MRP')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Average Sales per Item Type
plt.figure(figsize=(12, 6))
item_sales = df.groupby('Item_Type')['Item_Outlet_Sales'].mean().sort_values()
sns.barplot(x=item_sales.values, y=item_sales.index)
plt.title('Average Sales by Item Type')
plt.xlabel('Average Sales')
plt.ylabel('Item Type')
plt.tight_layout()
plt.show()
```

```python
# Plot 5: Sales by Outlet Type
plt.figure(figsize=(9, 6))
sns.boxplot(data=df, x='Outlet_Type', y='Item_Outlet_Sales')
plt.title('Sales Distribution by Outlet Type')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
# Plot 6: Outlet Size vs Sales
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x='Outlet_Size', y='Item_Outlet_Sales', palette="pastel")
plt.title('Sales by Outlet Size')
plt.tight_layout()
plt.show()
```

```python
# Plot 7: Sales by Outlet Location Type
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x='Outlet_Location_Type', y='Item_Outlet_Sales', palette='muted')
plt.title('Sales by Outlet Location Type')
plt.tight_layout()
plt.show()
```

```python
# Plot 8: Fat Content vs Sales
plt.figure(figsize=(9, 5))
sns.violinplot(data=df, x='Item_Fat_Content', y='Item_Outlet_Sales', palette='Set2')
plt.title('Sales by Fat Content')
plt.tight_layout()
plt.show()
```

```python
# Plot 9: Years of Outlet Operation vs Sales
plt.figure(figsize=(9, 5))
sns.barplot(data=df, x='Outlet_Establishment_Year', y='Item_Outlet_Sales', ci=None)
plt.title('Average Sales by Outlet Establishment Year')
plt.tight_layout()
plt.show()
```
